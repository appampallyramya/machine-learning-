```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
data=pd.read_csv('linear regression.csv')
```

```python
df=pd.read_csv("linear regression.csv")
```

```python
df.describe()
```

```python
sns.boxplot(x='radio',y='sales', data=df)
```

```python
plt.scatter(df['newspaper'],df['sales'])
plt.xlabel('newspaper')
plt.ylabel('sales')
plt.show()
```

```python
count,bin_edges=np.histogram(df['TV'])
df['TV'].plot(kind='hist',xticks=bin_edges)
plt.xlabel('value of TV')
plt.ylabel('number of sales')
plt.grid()
plt.show()
```

```python
sns.heatmap(df.isnull())
plt.show()
```

```python
correlation_matrix=df.corr()
sns.heatmap(correlation_matrix,annot=True)
plt.show()
```

```python
sns.regplot(x='newspaper',y='sales', data=df)
```
